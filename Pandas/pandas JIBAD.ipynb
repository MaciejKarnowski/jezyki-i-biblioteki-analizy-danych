```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1=pd.read_csv("Plant_1_Generation_Data.csv")
```

```python
df2=df=pd.read_csv("Plant_2_Generation_Data.csv")
```

```python
df1['DATE_TIME'] = pd.to_datetime(df1['DATE_TIME'])
df1
```

```python
df2['DATE_TIME'] = pd.to_datetime(df2['DATE_TIME'])
df2
```

```python
df1.isnull
```

```python
df=pd.concat([df1,df2])
```

```python
df=df.dropna()
```

```python
df['timestamp'] = pd.to_datetime(df['DATE_TIME'])

# extract the hour and minute from the timestamp column
df['day'] = df['timestamp'].dt.day
df['hour'] = df['timestamp'].dt.hour
df['minute'] = df['timestamp'].dt.minute
df['Week_Number'] = pd.to_datetime(df['timestamp']).dt.isocalendar().week
print(df)
```

```python
df.dtypes
```

```python
generator='1BY6WEcLGh8j5v7'

def generate_plot(df,week, generator_name):
    df_plot=df[df["Week_Number"]==week].reset_index()
    df1=df_plot.groupby(['day','hour','minute'])['AC_POWER'].mean().reset_index(name ='mean')
    df_plot=df_plot[df_plot['SOURCE_KEY']==generator_name]
    df_plot = pd.merge(df_plot, df1,  how='left', left_on=['hour','day','minute'], right_on = ['hour','day','minute'])
    df_plot.plot("DATE_TIME",["AC_POWER","mean"],rot=60)
```

```python
generate_plot(df,21,generator)
```

```python
df1=df.groupby(['DATE_TIME'])['AC_POWER'].mean().reset_index(name ='mean')
df1['4th_quantile']=df1["mean"]*0.8

df2= pd.merge(df, df1,  how='left', left_on=["DATE_TIME"], right_on = ['DATE_TIME'])

df_final=df2.where(df2["AC_POWER"]<df2["4th_quantile"]).dropna().reset_index(drop=True)
df_final = df_final.groupby(['SOURCE_KEY'])['SOURCE_KEY'].count().reset_index(name ='count').sort_values(by=['count'],ascending=False)

df_final2= df.groupby(['SOURCE_KEY'])['SOURCE_KEY'].count().reset_index(name ='count').sort_values(by=['count'],ascending=False)

df2= pd.merge(df_final, df_final2,  how='left', left_on=["SOURCE_KEY"], right_on = ['SOURCE_KEY'])

df2["percentage"]=round(df2["count_x"]/df2["count_y"]*100,2)
print(df2)
```
